```python
import nltk 
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt') 
nltk.download('stopwords') 
stop_words = set(stopwords.words('portuguese'))
```

```python
df = pd.read_excel('base.xlsx', index_col=0)
```

```python
# Esta fun칞칚o remove os dados que n칚o ser칚o utilizados para a an치lise 
def cleanData(df):
    # Remove as aspas dos nomes das colunas
    df = df.rename(columns=lambda x: x.replace('"', ''))
    # Remove coment치rios feito pelo btg
    dfWithoutAutor = df[df["autor"] != "btgpactual"]
    # Remove coment치rios nulos
    dfWithoutNull = dfWithoutAutor[dfWithoutAutor["autor"].notnull()]
    # Deixando apenas as intera칞칫es do tipo coment치rio e resposta
    dfFinal = dfWithoutNull.loc[(dfWithoutNull['tipoInteracao'] == 'coment치rio') | (dfWithoutNull['tipoInteracao'] == 'resposta')]
    return dfFinal
    
```

```python
def preprocessing_coments(frase):
    palavras = frase.split() #divide as frases em palavras
    filtered_words = [word for word in palavras if word.lower() not in stop_words and not word.startswith('@') and word.isalpha()] #coloca todas as letras em min칰sculas (lower) e remove as stop words
    tokens = word_tokenize(" ".join(filtered_words)) #tokeniza as palavras e organiza cada coment치rio em uma lista de palavras
    return tokens
```

```python

#Fun칞칚o para rodar nosso modelo de bag of words, essa fun칞칚o recebe um dataframe
def bag_of_words(base):
    #limpeza da base de dados
    base1 = cleanData(base)
    #tokeniza칞칚o da coluna de textos
    base1['Frases_sem_stop_words'] = base1['texto'].apply(preprocessing_coments)
    return base1
    
```

```python
base_nova = bag_of_words(df)
```
