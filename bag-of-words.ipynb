```python
import pandas as pd
```

```python
df = pd.read_csv("./base/base.csv", encoding='UTF-8')
```

```python
# this function cleans up data that will not be used in the model
def cleanData(df):
    #remove comments made by btg
    dfWithoutAutor = df[df["autor"] != "btgpactual"]
    #remove comments with null authors
    dfWithoutNull = dfWithoutAutor[dfWithoutAutor['autor'].notnull()]
    #leave only the comment texts
    dfFinal = dfWithoutNull[dfWithoutNull["tipoInteracao"] == "comment"]
    return dfFinal
    
```

```python
df1 = cleanData(df)
```
