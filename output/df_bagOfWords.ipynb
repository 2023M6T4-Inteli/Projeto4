```python
from keras.preprocessing.text import Tokenizer
# Realizando o Bag of Words


## Step 1: Determine the Vocabulary
tokenizer = Tokenizer()
tokenizer.fit_on_texts(df['Frases_sem_stop_words'])
print(f'Vocabulary: {list(tokenizer.word_index.keys())}')

## Step 2: Count
vectors = tokenizer.texts_to_matrix(df['Frases_sem_stop_words'], mode='count')
print(vectors)
```

```python
# Função para retornar um dataframe com as palavras mais recorrentes
def bow_dataframe(inputUser):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(inputUser)
    wordCount = tokenizer.word_counts    # Com essa variavel é possível ver as palavras e contar sua frequência 
    dfCountBoW = pd.DataFrame(list(wordCount.items()))
    dfCountBoW.rename(columns={0: "Palavra", 1:"Frequência"}, inplace=True)
    final_df = dfCountBoW.sort_values(by=['Frequência'], ascending=False)
    return final_df
```

```python
teste = bow_dataframe(df['Frases_sem_stop_words'])
teste
```
