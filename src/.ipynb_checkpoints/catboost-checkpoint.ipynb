### CatBoost

O CatBoost é um algoritmo de classificação que se destaca pela sua capacidade de lidar com dados categóricos, fornecendo resultados precisos e eficientes em treinamento. Com recursos avançados de tratamento de dados e prevenção de sobreajuste, o CatBoost é uma escolha popular em diversas aplicações, desde análise de fraudes até sistemas de recomendação.

```python
# Importando as bibliotecas
import catboost
import pandas as pd
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import recall_scor
from sklearn.metrics import precision_score
from sklearn.metrics import accuracy_score
```

```python
# Lista de arquivos
arquivos = ['../output/word2vec_corpus.csv', '../output/word2vec_cbow.csv']
```

```python
# Loop para ler e treinar o modelo para cada arquivo
for arquivo in arquivos:
    # Lendo os dados
    df = pd.read_csv(arquivo)

    # Separando os dados em features e target
    X = df.drop(['Frase', 'sentimento'], axis=1)
    y = df['sentimento']

    # Dividindo em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
    # Treinando o modelo
    model = CatBoostClassifier(iterations=200, learning_rate=0.1)
    model.fit(X_train, y_train)
```

```python
    # Avaliando o modelo no conjunto de teste
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Acurácia para o arquivo {arquivo}: {accuracy:.2f}')
    print('---')
```

```python
print(f'Acurácia para o arquivo ../output/word2vec_corpus.csv: {accuracy:.2f}')
```

### Matriz de confusão

A matriz de confusão tem como objetivo, comparar as predições do modelo com o dataset real. De tal forma, nesse problema de multiclassificação é possível ter um panorama geral das respostas e se o modelo tem algum viés ou não.

```python
# Função para plotar matriz de confusão
def plot_confusion_matrix(y_true, y_pred, title):
    confusion_matrix_corpus = confusion_matrix(y_true, y_pred, normalize='true')
    cm_corpus = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_corpus,
                                       display_labels=['Negativo', 'Neutro', 'Positivo'])
    cm_corpus.plot()
    cm_corpus.ax_.set(title=title,
                      xlabel='Sentimentos previstos',
                      ylabel='Sentimentos reais')
    plt.show()
```

```python
# Lendo os dados - word2vec_corpus.csv
df_corpus = pd.read_csv('../output/word2vec_corpus.csv')
```

```python
# Separando os dados em features e target
X_corpus = df_corpus.drop(['Frase', 'sentimento'], axis=1)
y_corpus = df_corpus['sentimento']

# Dividindo em treino e teste
X_train_corpus, X_test_corpus, y_train_corpus, y_test_corpus = train_test_split(X_corpus, y_corpus, test_size=0.2)

# Treinando o modelo - word2vec_corpus.csv
model_corpus = CatBoostClassifier(iterations=200, learning_rate=0.1)
model_corpus.fit(X_train_corpus, y_train_corpus)

# Avaliando o modelo no conjunto de teste - word2vec_corpus.csv
y_pred_corpus = model_corpus.predict(X_test_corpus)
plot_confusion_matrix(y_test_corpus, y_pred_corpus, 'Matriz de confusão para word2vec_corpus.csv')
```

```python
# Lendo os dados - word2vec_cbow.csv
df_cbow = pd.read_csv('../output/word2vec_cbow.csv')
```

```python
# Separando os dados em features e target
X_cbow = df_cbow.drop(['Frase', 'sentimento'], axis=1)
y_cbow = df_cbow['sentimento']

# Dividindo em treino e teste
X_train_cbow, X_test_cbow, y_train_cbow, y_test_cbow = train_test_split(X_cbow, y_cbow, test_size=0.2)

# Treinando o modelo - word2vec_cbow.csv
model_cbow = CatBoostClassifier(iterations=200, learning_rate=0.1)
model_cbow.fit(X_train_cbow, y_train_cbow)

# Avaliando o modelo no conjunto de teste - word2vec_cbow.csv
y_pred_cbow = model_cbow.predict(X_test_cbow)
plot_confusion_matrix(y_test_cbow, y_pred_cbow, 'Matriz de confusão para word2vec_cbow.csv')
```

### Recall


Recall é uma métrica de avaliação utilizada em problemas de classificação para medir a capacidade de um modelo em identificar corretamente os exemplos positivos, ou seja, a proporção de casos positivos que foram corretamente identificados pelo modelo em relação ao total de casos positivos.

```python
# Função para calcular o recall
def calculate_recall(X, y, model):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    recall = recall_score(y_test, y_pred, average='macro')
    return recall
```

```python
# Lendo os dados - word2vec_corpus.csv
df_corpus = pd.read_csv('../output/word2vec_corpus.csv')
```

```python
# Separando os dados em features e target
X_corpus = df_corpus.drop(['Frase', 'sentimento'], axis=1)
y_corpus = df_corpus['sentimento']

# Criando o modelo - word2vec_corpus.csv
model_corpus = CatBoostClassifier(iterations=200, learning_rate=0.1)
```

```python
# Calculando o recall - word2vec_corpus.csv
recall_corpus = calculate_recall(X_corpus, y_corpus, model_corpus)
print(f'Recall para word2vec_corpus.csv: {recall_corpus:.2f}')
```

```python
# Lendo os dados - word2vec_cbow.csv
df_cbow = pd.read_csv('../output/word2vec_cbow.csv')
```

```python
# Separando os dados em features e target
X_cbow = df_cbow.drop(['Frase', 'sentimento'], axis=1)
y_cbow = df_cbow['sentimento']

# Criando o modelo - word2vec_cbow.csv
model_cbow = CatBoostClassifier(iterations=200, learning_rate=0.1)
```

```python
# Calculando o recall - word2vec_cbow.csv
recall_cbow = calculate_recall(X_cbow, y_cbow, model_cbow)
print(f'Recall para word2vec_cbow.csv: {recall_cbow:.2f}')
```

Dado que os valores de recall para ambos os arquivos são 58%, não há diferenças significativas na capacidade de recuperação dos exemplos positivos entre os dois modelos

### Precisão


Precisão é uma métrica de avaliação utilizada em problemas de classificação para medir a proporção de exemplos classificados corretamente como positivos em relação ao total de exemplos classificados como positivos pelo modelo, ou seja, a capacidade do modelo em fazer previsões corretas dos casos positivos.

```python
# Função para calcular a precisão
def calculate_precision(X, y, model):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    precision = precision_score(y_test, y_pred, average='macro')
    return precision
```

```python
# Lendo os dados - word2vec_corpus.csv
df_corpus = pd.read_csv('../output/word2vec_corpus.csv')
```

```python
# Separando os dados em features e target
X_corpus = df_corpus.drop(['Frase', 'sentimento'], axis=1)
y_corpus = df_corpus['sentimento']

# Criando o modelo - word2vec_corpus.csv
model_corpus = CatBoostClassifier(iterations=200, learning_rate=0.1)
```

```python
# Calculando a precisão - word2vec_corpus.csv
precision_corpus = calculate_precision(X_corpus, y_corpus, model_corpus)
print(f'Precisão para word2vec_corpus.csv: {precision_corpus:.2f}')
```

```python
# Lendo os dados - word2vec_cbow.csv
df_cbow = pd.read_csv('../output/word2vec_cbow.csv')
```

```python
# Separando os dados em features e target
X_cbow = df_cbow.drop(['Frase', 'sentimento'], axis=1)
y_cbow = df_cbow['sentimento']

# Criando o modelo - word2vec_cbow.csv
model_cbow = CatBoostClassifier(iterations=200, learning_rate=0.1)
```

```python
# Calculando a precisão - word2vec_cbow.csv
precision_cbow = calculate_precision(X_cbow, y_cbow, model_cbow)
print(f'Precisão para word2vec_cbow.csv: {precision_cbow:.2f}')
```

A precisão obtida para o conjunto de dados "word2vec_cbow.csv" é de 0.60, o que significa que o modelo classificou corretamente 60% dos exemplos positivos em relação ao total de exemplos classificados como positivos.

Já a precisão obtida para o conjunto de dados "word2vec_corpus.csv" é de 0.63, indicando que o modelo classificou corretamente 63% dos exemplos positivos em relação ao total de exemplos classificados como positivos.

Portanto, com base nas informações fornecidas, podemos observar que o modelo teve um desempenho ligeiramente melhor no conjunto de dados "word2vec_corpus.csv" em comparação ao conjunto de dados "word2vec_cbow.csv", com uma precisão mais alta de 0.63 em comparação a 0.60.
