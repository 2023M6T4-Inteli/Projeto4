### Word2Vec com o corpus

Nesta etapa ser√° realizado o processamento a partir do zero, assim ser√° poss√≠vel ter dois modelos diferentes permitindo uma melhor oportunidade para realizar os testes.

```python
# Abrindo a base j√° pr√©-processada 
df = pd.read_csv('../output/sentimento_revisto.csv')
df
```

```python
# Definindo qual coluna ser√° utilizada para treinar o modelo 
input_training = df['texto_tokenizado2'] 
```

```python
# Treinando o word2vec com o corpus
model_corpus = Word2Vec(input_training, min_count=1)
```

```python
# Fun√ß√£o que soma todas as palavras e cria as colunas
def create_sentence_vector(model, cleaned_df):
    sentence_table = []
    for sentence in cleaned_df['texto_tokenizado']:
        word_vectors = [model.wv[word] for word in sentence if word in model.wv]
        if len(word_vectors) > 0:
            sentence_vector = sum(word_vectors) / len(word_vectors)
        else:
            sentence_vector = [None] * 100  # Cria uma lista de 100 elementos None
        sentence_table.append((sentence, *sentence_vector[:50]))  # Adiciona apenas os primeiros 50 elementos do vetor

    column_labels = ['Frase']
    for i in range(50):
        column_labels.append(f'Vetor{i+1}')
    df_vec = pd.DataFrame(sentence_table, columns=column_labels)

#     cleaned_df["sentimentoNumerico"] = cleaned_df["sentimento"].replace({'NEGATIVE': -1, 'POSITIVE': 1, 'NEUTRAL': 0})

#     # Definir o √≠ndice do DataFrame df_vec como o mesmo √≠ndice de df_processada['sentimentoNumerico']
#     df_vec.set_index(cleaned_df["sentimentoNumerico"].index, inplace=True)

    df_vec['sentimento'] = cleaned_df["sentimento_revisto"]
    df_vec = df_vec.dropna()

    return df_vec
```

```python
# Criando os vetores
df_corpus = create_sentence_vector(model_corpus, df)
df_corpus
```

```python
# Exportando o csv com o resultado da aplica√ß√£o do word2vec com os vetores do corpus
df_corpus.to_csv('..\\output\\word2vec_corpus.csv',encoding='utf-8', index=False, header=True, )
```
