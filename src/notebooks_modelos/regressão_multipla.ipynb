```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import linear_model
import numpy as np 
```

```python
df_corpus = pd.read_csv('../output/word2vec_corpus.csv')
df_corpus.head(5)
```

```python
# label encoding nas colunas
label_encoder = LabelEncoder()
target_var = df_corpus['sentimento']
encoded_labels = label_encoder.fit_transform(target_var)
encoded_labels
```

```python
# Definindo as variaveis de entrada e as de saida 
df_input = df_corpus.iloc[:,1 :51]
df_input
```

```python
df_target = df_corpus['sentimento']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_input, df_target, train_size = 0.8)
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Create an instance of the LinearRegression class
reg = LinearRegression()
 
# Fit the model to the data
reg.fit(x_train, y_train)
 
# Print the coefficients of the model
print(reg.coef_)
```

```python
predictions = reg.predict(x_test)
```

```python
# model evaluation
print('mean_squared_error : ', mean_squared_error(y_test, predictions))
print('mean_absolute_error : ', mean_absolute_error(y_test, predictions))
```

```python
r_squared = reg.score(x_test, y_test)
r_squared
```
