```python
cbow = '../base/cbow_s50.txt'
```

```python
from scipy.spatial.distance import cosine

from gensim.models import KeyedVectors

model_cbow = KeyedVectors.load_word2vec_format(cbow)
```

```python
# Testando o word2vec
wordvec_word1 = model_cbow['banco']

wordvec_word1
```

```python
def word2vec(tokenized_test):
    array_full = []
    for x in tokenized_test:
        vector = model_cbow[x]
        array_full.append(vector)
    final_output = sum(array_full)  
    return final_output
        
```

```python
df['veczada'] = df['texto_tokenizado'].apply(word2vec)
```
